```python
#want to do singular value decomposition
#should perform well with the sparse dataset
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# define the pipeline
#after evaluation n_components=10 yielded the best result
# for i in range(1,20):
#     print(i*5)
#     steps = [('scaler', StandardScaler()), ('svd', TruncatedSVD(n_components=i*5)), ('m', LogisticRegression(max_iter=20000))]
#     svd = Pipeline(steps=steps)
#     scores = evaluate(svd)
#     print('root mean squared error: ', scores)

steps = [('svd', TruncatedSVD(n_components=10)), ('m', LogisticRegression(max_iter=50000))]
#steps = [('scaler', StandardScaler()), ('svd', TruncatedSVD(n_components=40)), ('m', LogisticRegression(max_iter=20000))]
svd = Pipeline(steps=steps)
scores = evaluate(svd)
print('root mean squared error: ', scores)

#print out the validation
#print(y_valid, len(y_valid), type(y_valid))
```
